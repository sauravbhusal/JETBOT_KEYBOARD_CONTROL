```python
#importing modules
import cv2
import ipywidgets.widgets as widgets
import traitlets
from jetbot import bgr8_to_jpeg
image = widgets.Image(format='jpeg', width=300, height=300)

from jetbot import Robot
import time
import requests
robot = Robot()

from jetbot.camera.opencv_gst_camera import OpenCvGstCamera
Camera = OpenCvGstCamera()
```

```python
#streaming_live_from_camera

camera_link = traitlets.dlink((Camera, 'value'), (image, 'value'), transform=bgr8_to_jpeg)
display(image)
```

```python
#keyboard_control

while True:
    res = requests.get('ip_address_where_server_streamed_the_value')
    print ('val = ',res.text)
    
    if res.text == '8':
        robot.backward(0.2)
        
    elif res.text == '4':
        robot.left(0.2)
        
    elif res.text == '6':
        robot.right(0.2)
        
    elif res.text == '2':
        robot.forward(0.2)

    elif res.text == '1':
        robot.backward(1)
        
    elif res.text == '0':
        robot.stop()
```
